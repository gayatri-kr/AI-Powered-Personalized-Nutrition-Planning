**Dietary**

* Files: DR1IFF_L.xpt, DRXFCD_L.xpt
* Variables Information
https://wwwn.cdc.gov/nchs/nhanes/search/variablelist.aspx?Component=Dietary&Cycle=2021-2023
* Selected variables: 'SEQN','DR1_030Z','DR1CCMTX','DR1IFDCD','DR1IGRMS','DR1IKCAL','DR1IPROT','DR1ICARB','DR1ISUGR','DR1IFIBE','DR1ITFAT','DR1ICHOL'

* food and nutrients

```python
dfn = pd.read_sas("DR1IFF_L.xpt")[['SEQN','DR1_030Z','DR1CCMTX','DR1IFDCD','DR1IGRMS','DR1IKCAL','DR1IPROT','DR1ICARB','DR1ISUGR','DR1IFIBE','DR1ITFAT','DR1ICHOL']]
```

```python
dfn.head()
```

```python
# Handling floating point artifacts
dfn = dfn.applymap(lambda x: 0 if isinstance(x, (int, float)) and abs(x) < 1e-10 else x)
dfn.head()
```

```python
# Rename Cols
dfn.columns = ['SEQN','Occasion','DietType','FoodCode','Grams','Energy','Proteins','Carbs','Sugars','Fiber','Fat','Cholestrol']
```

```python
dfn.head()
```

```python
# Converting Codes to int
dfn.SEQN = dfn.SEQN.astype(int)
dfn.DietType = dfn.DietType.astype(int)
dfn.Occasion = dfn.Occasion.astype(int)
dfn.FoodCode = dfn.FoodCode.astype(int)
dfn.head()
```

* Working on Occassion, DietType & FoodNames

```python
# Mappings from above url
combination_food_type = {
    0: "Non-combination food",
    1: "Beverage w/ additions",
    2: "Cereal w/ additions",
    3: "Bread/baked products w/ additions",
    4: "Salad",
    5: "Sandwiches",
    6: "Soup",
    7: "Frozen meals",
    8: "Ice cream/frozen yogurt w/ additions",
    9: "Dried beans and vegetable w/ additions",
    10: "Fruit w/ additions",
    11: "Tortilla products",
    12: "Meat, poultry, fish",
    13: "Lunchables®",
    14: "Chips w/ additions",
    15: "Baby Toddler Food and Infant Formula",
    90: "Other mixtures",
    ".": "Missing"
}
eating_occasion = {
    1: "Breakfast",
    2: "Lunch",
    3: "Dinner",
    4: "Supper",
    5: "Brunch",
    6: "Snack",
    7: "Drink",
    8: "Infant feeding",
    9: "Extended consumption",
    10: "Desayuno",
    11: "Almuerzo",
    12: "Comida",
    13: "Merienda",
    14: "Cena",
    15: "Entre comida",
    16: "Botana",
    17: "Bocadillo",
    18: "Tentempie",
    19: "Bebida",
    91: "Other",
    99: "Don’t know",
    ".": "Missing"
}
```

```python
import warnings
warnings.filterwarnings("ignore")

dfn.DietType = dfn.DietType.replace(combination_food_type)
dfn.Occasion = dfn.Occasion.replace(eating_occasion)
```

```python
dfn.head()
```

**USDA** food codes given along with description

```python
foodinfo = pd.read_sas("DRXFCD_L.xpt")[['DRXFDCD','DRXFCSD']]
```

```python
foodinfo
```

```python
foodinfo.columns = ['FoodCode','FoodDescription']
```

```python
foodinfo.FoodCode = foodinfo.FoodCode.astype(int)
```

```python
foodinfo
```

```python
# Cleaning Description
foodinfo.FoodDescription = foodinfo.FoodDescription.apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)
```

```python
foodinfo.head()
```

* Mapping Food Description to above diet data based on Food Code

```python
dfn = dfn.merge(foodinfo, on='FoodCode', how='left')
```

```python
dfn.head()
```

```python
# De-Duplication
dfn = dfn.drop_duplicates().reset_index(drop=True)
dfn.shape
```

```python
# grouping by Diet & Occassion
dfn = (
    dfn.groupby(['SEQN', 'Occasion', 'DietType','FoodDescription'])
       .agg({'Grams':'sum','Energy':'sum','Proteins':'sum','Carbs':'sum','Sugars':'sum','Fiber':'sum','Fat':'sum','Cholestrol':'sum'})
       .reset_index()
)
dfn.head()
```

```python
dfn.shape
```

```python
dfn.DietType.unique()
```

* Mental Health

```python
# Over The Last Two Weeks
mh = pd.read_sas("DPQ_L.xpt")
```

```python
mh
```

```python
mh.SEQN = mh.SEQN.astype(int)
mh.head()
```

```python
# Handling floating point artifacts
mh = mh.applymap(lambda x: 0 if isinstance(x, (int, float)) and abs(x) < 1e-10 else x)
mh.head()
```

* Considering above into three major categories upon observation

```python
stress_cols = ['DPQ010', 'DPQ050', 'DPQ070', 'DPQ080', 'DPQ100']
sleep_cols = ['DPQ030']
mood_cols = ['DPQ020', 'DPQ040', 'DPQ060', 'DPQ090']
```

```python
mh['Stress'] = round(mh[stress_cols].mean(axis=1))
mh['Sleep'] = round(mh[sleep_cols].mean(axis=1))
mh['Mood'] = round(mh[mood_cols].mean(axis=1))
```

```python
mh = mh[['SEQN', 'Stress', 'Sleep', 'Mood']]
```

* Col data validation

```python
for col in mh.columns[1:]:
    print(col, mh[col].unique())
```

* According to variable description in above url we have 0,1,2,3,7,9 only
    - replacing others to close value

```python
mh.replace({4.0:7.0, 5.0:7.0}, inplace=True)
```

```python
for col in mh.columns[1:]:
    print(col, mh[col].unique())
```

```python
mh.head()
```

**Final Merge**

```python
demo.shape, bmx.shape, bp.shape, ch.shape, db.shape, mh.shape, dfn.shape
```

```python
# Merge Commands: Picking SEQN Matched rows only -> Inner
data = demo.copy()
data = pd.merge(data, bmx, on='SEQN', how='inner')
data = pd.merge(data, bp, on='SEQN', how='inner')
data = pd.merge(data, ch, on='SEQN', how='inner')
data = pd.merge(data, db, on='SEQN', how='inner')
data = pd.merge(data, mh, on='SEQN', how='inner')
data = pd.merge(data, dfn, on='SEQN', how='inner')

print(f'combined dataset size : {data.shape}')
```
