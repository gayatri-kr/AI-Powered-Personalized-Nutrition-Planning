```python
# Modules
import pandas as pd
import numpy as np

import joblib

import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.models import load_model

AUTOTUNE = tf.data.AUTOTUNE

import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading Saved Model & Its Supporting objects
algo = load_model("/kaggle/input/fusion-predictionfiles/fusion.keras")

# Scalers
shm = joblib.load("/kaggle/input/fusion-predictionfiles/shm.pkl")
smh = joblib.load("/kaggle/input/fusion-predictionfiles/smh.pkl")
sg = joblib.load("/kaggle/input/fusion-predictionfiles/sg.pkl")
sp = joblib.load("/kaggle/input/fusion-predictionfiles/sp.pkl")
sy = joblib.load("/kaggle/input/fusion-predictionfiles/sy.pkl")
```

**UInputs**

```python
print("Model Trained on Below Input Features:")
print("Gender, Age, BMI, Bp, Cholestrol, Diabetes, Stress, Sleep, Mood, Snp_Seq, Protein Folds")
print()
print("Enter Above Values one by one:")
print()
print("############ Health Metrics ################")
g = input("Enter Person Gender (Male/Female):")
a = eval(input("Enter Person Age:"))
b = eval(input('Enter Person BMI (Kg/m2):'))
sb = eval(input("Enter Systolic_bp(mmHg):"))
db = eval(input("Enter Diastolic_bp(mmHg):"))
ch = eval(input("Enter Cholesterol(mg/dL):"))
d = input('Enter Person Diabetes Status (Yes, No, Borderline):')
print()
print("############ Mental Health Metrics ################")
print("Mental Health Status:(Not at all, Several days, More than half the days, Nearly every day, Refused, Don't know):")
st = input("Enter For Stress:")
sl = input("Enter For Sleep:")
mo = input("Enter For Mood:")
print()
print("############ 100 val Snp_Seq ################")
snp = input("Enter SnpSeq List:")
print()
print("############ 100 val ProteinFold ################")
pf = input("Enter Protein Fold Seq List:")

print()
inp = [g,a,b,sb,db,ch,d,st,sl,mo,snp,pf]
# Converting to Df
row = pd.DataFrame([inp], columns = ['Gender', 'Age', 'BMI', 'Systolic_bp', 'Diastolic_bp', 'Cholesterol',
                                   'Diabetes', 'Stress', 'Sleep', 'Mood', 'Snp_Sequence',
                                   'Protein_Sequence_Folds'])

print("Given Input Data:")
display(row)
```

* Preparation

```python
row.Gender.replace({'Male':1, 'Female':2}, inplace=True)
row.Diabetes.replace({'Yes':1, 'No':2, 'Borderline':3}, inplace=True)
row.replace({'Not at all':0, 'Several days':1, 'More than half the days':2, 'Nearly every day':3, 
                                        'Refused':7, "Don't know":9}, inplace=True)
# Health Metrics
health_cols = ['Gender', 'Age', 'BMI', 'Systolic_bp', 'Diastolic_bp', 'Cholesterol', 'Diabetes']
xhm = row[health_cols].values.astype(np.float32)
# Mental Health
mentalhealth_cols = ['Stress','Sleep','Mood']
xmh = row[mentalhealth_cols].values.reshape(len(row), len(mentalhealth_cols), 1).astype(np.float32)  # (N, 3, 1)
# SNP
import ast
row['Snp_Sequence'] = row['Snp_Sequence'].apply(lambda x: np.array(ast.literal_eval(x), dtype=np.int8))
xgen = np.stack(row['Snp_Sequence'].values)
# Protein Folds
row['Protein_Sequence_Folds'] = row['Protein_Sequence_Folds'].apply(lambda x: np.array(ast.literal_eval(x), dtype=np.float32))
xpro = np.stack(row['Protein_Sequence_Folds'].values)

print("Data Shapes:")
print("X_healthmetrics", xhm.shape)
print("X_mentalhealth", xmh.shape)
print("X_genomic", xgen.shape)
print("X_protein", xpro.shape)
```

* Pre-Processing

```python
xhm_scaled = shm.transform(xhm)
xmh_scaled = smh.transform(xmh.reshape(-1, xmh.shape[-1]))
xgen_scaled = sg.transform(xgen)
xpro_scaled = sp.transform(xpro)

# Reshaping Mental health data
xmh_scaled = xmh_scaled.reshape(xmh.shape)
```

```python
test_ds = tf.data.Dataset.from_tensor_slices((
    {
        "healthmetrics_input": xhm_scaled.astype(np.float32),
        "mentalhealth_input": xmh_scaled.astype(np.float32),
        "genomic_input": xgen_scaled.astype(np.int32),
        "protein_input": xpro_scaled.astype(np.float32)
    }
)).batch(64).prefetch(AUTOTUNE)
```

```python
pred_scaled = algo.predict(test_ds)['nutrition_targets']
pred_real = sy.inverse_transform(pred_scaled)[0]
y_cols = ['Grams (g)','Energy (kcal)','Proteins (g)','Carbs (g)','Sugars (g)','Fiber (g)','Fat (g)','Cholesterol (mg)']
print("Predicted Nutrients for the Day:")
for name, val in zip(y_cols, pred_real):
    print(f"{name:12s}: {val:8.2f}")
```
